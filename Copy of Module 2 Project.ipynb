```python
# Handle missing values
# We used placeholders for categorical/text data and median for numerical data
if 'df' in locals():
    df['description'] = df['description'].fillna('No description available')
    df['age_certification'] = df['age_certification'].fillna('N/A')
    df['seasons'] = df['seasons'].fillna(0)
    df['imdb_id'] = df['imdb_id'].fillna('N/A')

    # Fill numerical columns with median
    numeric_cols = ['imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']
    for col in numeric_cols:
        df[col] = df[col].fillna(df[col].median())

    print("NaN values have been replaced. Remaining nulls:")
    print(df.isnull().sum())
else:
    print("Warning: 'df' not found. ")
```

```python
# Properly clean the genres column by removing brackets and quotes
cleaned_genres = df.dropna(subset=['genres']).copy()
cleaned_genres['genres'] = cleaned_genres['genres'].str.strip("[]").str.replace("'", "").str.split(", ")

# Explode into individual rows
exploded_genres = cleaned_genres.explode('genres')

# Strip any remaining whitespace
exploded_genres['genres'] = exploded_genres['genres'].str.strip()

# Check the new top genres counts
top_genres_cleaned = exploded_genres['genres'].value_counts().head(10)
print("Cleaned Top 10 Genres:")
print(top_genres_cleaned)
```

# ***Content Growth Over Time***
**Line Chart**

```python
growth = df.groupby(['release_year', 'type']).size().reset_index(name='count')

plt.figure()
for t in growth['type'].unique():
    subset = growth[growth['type'] == t]
    plt.plot(subset['release_year'], subset['count'], label=t)

plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.title('Content Growth Over Time')
plt.legend()
plt.show()
```

### **Insights from Content Growth Over Time**

1.  **Exponential Growth:** There is a clear exponential increase in the number of titles released per year, particularly starting from the early 2000s and accelerating sharply after 2010. This aligns with the rise of digital production and streaming platforms.
2.  **Movie Dominance Across Eras:** Throughout history, the number of movies released annually has consistently surpassed the number of shows. However, both formats show a similar upward trajectory in recent decades.
3.  **Recent Peak:** The highest volume of content in the dataset comes from the last 5-10 years, indicating that the library is heavily weighted towards modern releases.
4.  **Historical Data Sparse:** Content from before 1970 is relatively sparse, representing a collection of classic cinema rather than high-volume annual production.

# ***IMDB Score vs Popularity***
**Scatter Plot**

```python
#removing the missing values
rating_df = df.dropna(subset=['imdb_score', 'tmdb_popularity'])

plt.figure()
plt.scatter(
    rating_df['imdb_score'],
    rating_df['tmdb_popularity'],
    s=rating_df['imdb_votes'] / 1000
)

plt.xlabel('IMDb Score')
plt.ylabel('TMDB Popularity')
plt.title('Ratings vs Popularity')
plt.show()
```

### **Insights from Ratings vs Popularity Scatter Plot**

1.  **Mainstream Cluster:** Most titles with a high number of votes (larger bubbles) are concentrated in the **6.5 to 8.5 IMDb score** range and have moderate to high popularity. These represent well-known mainstream successes.
2.  **Weak Correlation:** There isn't a strict linear relationship between rating and popularity. Many highly-rated titles (8.5+) have relatively low popularity scores, indicating they might be niche or 'hidden gems.'
3.  **Popularity Outliers:** Some titles exhibit exceptionally high TMDB popularity scores despite having average IMDb ratings (around 6.0-7.0). These are often trending new releases or viral content that captures immediate attention regardless of critical score.
4.  **Vote Volume vs. Score:** Larger bubbles (representing more votes) tend to shift towards higher IMDb scores, suggesting that the most-watched content generally maintains a certain standard of quality or broad appeal.

4.  **Rarity of Extremes:** Both critically acclaimed masterpieces (9.0+) and very poorly received titles (below 3.0) are relatively rare, representing the outliers of the content library.

# ***Country-wise Content Production***
**Bar Chart**

```python
countries = df.dropna(subset=['production_countries'])
countries = countries.assign(
    production_countries=countries['production_countries'].str.strip("[]'").str.split("', '")
)
countries = countries.explode('production_countries')

# Strip whitespace to handle potential empty strings correctly
countries['production_countries'] = countries['production_countries'].str.strip()

# Mapping ISO codes to full names
country_map = {
    'US': 'United States',
    'IN': 'India',
    'GB': 'United Kingdom',
    'FR': 'France',
    'JP': 'Japan',
    'ES': 'Spain',
    'KR': 'South Korea',
    'CA': 'Canada',
    'DE': 'Germany',
    'MX': 'Mexico',
    'BR': 'Brazil',
    'IT': 'Italy',
    'AU': 'Australia',
    'TR': 'Turkey',
    'PH': 'Philippines'
}

countries['production_countries'] = countries['production_countries'].replace(country_map)

# Filter out entries that are still empty strings
countries = countries[countries['production_countries'] != '']

top_countries = countries['production_countries'].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_countries.index[::-1], top_countries.values[::-1], color='lightblue')
plt.xlabel('Number of Titles')
plt.title('Top 10 Content Producing Countries (Full Names)')
plt.show()
```

### **Insights from Country-wise Content Production**

1.  **United States Lead:** The US dominates the dataset with over **5,300 titles**, highlighting its position as a global hub for entertainment content.
2.  **Major International Contributors:** **India (1,072)** and the **United Kingdom (928)** are the next most prolific producers, showcasing strong film and television industries.
3.  **Production Concentration:** The gap between the top producer (US) and the rest is massive, indicating a high level of concentration in content sourcing.
4.  **Diverse Geography:** The top 10 includes representation from North America, Asia (India, Japan), Europe (UK, France, Germany, Italy), and Oceania (Australia).
